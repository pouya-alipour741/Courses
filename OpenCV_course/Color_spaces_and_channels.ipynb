```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# color spaces

```python
img = cv.imread("assets/Photos/park.jpg")

cv.imshow("image", img)
cv.waitKey(0)
cv.destroyAllWindows()

##visual difference between BGR and RGB
plt.imshow(img)
plt.axis(False)
plt.show()
```

```python
#BGR to RGB
rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)

cv.imshow("rgb", rgb)
cv.waitKey(0)
cv.destroyAllWindows()

plt.imshow(rgb)
plt.axis(False)
plt.show()
```

```python
#BGR to HSV
hsv = cv.cvtColor(img, cv.COLOR_BGR2HSV)

cv.imshow("hsv", hsv)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#BGR to LAB
lab = cv.cvtColor(img, cv.COLOR_BGR2LAB)

cv.imshow("lab", lab)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
## note: can't convert hsv or lab to gray directly or vice versa. must convert to BGR first
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

cv.imshow("gray", gray)
cv.waitKey(0)
cv.destroyAllWindows()
```

# color channels

```python
b,g,r = cv.split(img)

cv.imshow("blue", b)  ##sky is white in this and that means high concentration of blue in this region
cv.imshow("green", g)
cv.imshow("red", r)
cv.waitKey(0)
cv.destroyAllWindows()

print(img.shape)
print(b.shape)

merged = cv.merge([b,g,r])
cv.imshow("merged image", merged)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
## show images in separate blue, green, red format
blank = np.zeros(img.shape[:2], dtype="uint8")

blue = cv.merge([b,blank,blank])
green = cv.merge([blank,g,blank])
red = cv.merge([blank,blank,r])

cv.imshow("blue", blue)  
cv.imshow("green", green)
cv.imshow("red", red)
cv.waitKey(0)
cv.destroyAllWindows()
```
