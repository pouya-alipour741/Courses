```python
import pandas as pd
import numpy as np

from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer, KNNImputer
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/titanic_train.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df = df.loc[:,["Pclass","Age","Survived"]]
df
```

```python
# filter for only na values
na_df = df["Age"].isna()
na_df
```

```python
# try to predict n/a with regression model independent of our model
imputer_it = IterativeImputer()
imputer_it.fit_transform(df)
```

```python
# newly generated values for n/a data
pd.DataFrame(imputer_it.fit_transform(df), columns=df.columns)[na_df]
```

```python
# n neighbour is our hyper parameter
impute_knn = KNNImputer(n_neighbors=2)
impute_knn.fit_transform(df)
```

```python
# newly generated values for n/a data with knn imputer
pd.DataFrame(impute_knn.fit_transform(df), columns=df.columns)[na_df]
```
