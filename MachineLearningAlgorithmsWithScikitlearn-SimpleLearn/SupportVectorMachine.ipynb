## classify aligators vs crocodiles

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
X, y = make_blobs(n_samples=1000, centers=2, random_state=20)
X
```

```python
clf = svm.SVC(kernel="linear",C=1)
clf.fit(X, y)
```

```python
X[:2]
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=30, cmap=plt.cm.Paired)
```

```python
y_pred = clf.predict(X[:5]) #our y has1d shape so we use X(2d) here for learning purposes
y_pred
```

```python
class_names = np.array(["aligator", "crocodile"])
```

```python
labels = class_names[y_pred]
labels
```

```python
clf.decision_function(X[:5])
```

```python
clf.classes_
```

```python
##same as above 
clf = svm.SVC(kernel="linear",C=1000)
clf.fit(X,y)

plt.figure(figsize=(15,10))
ax = plt.scatter(X[:,0],X[:,1],c=y, cmap=plt.cm.Paired)
# plt.plot([1,2,3],[4,5,6])
# ax = plt.gca()
xlim = plt.xlim()
ylim = plt.ylim()

#create grid and eval model
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)

YY, XX  = np.meshgrid(xx, yy)

xy = np.vstack([np.ravel(XX),np.ravel(YY)]).T
xy.shape

z = clf.decision_function(xy).reshape(XX.shape)

#plot decision boundary and margins
plt.contour(XX, YY, z, colors="k", levels=[-1,0,1], alpha=0.5, linestyles=["--","-","--"])
##plot support vectors
plt.scatter(clf.support_vectors_[0], clf.support_vectors_[1],s=100, linewidths=1, facecolors='#17becf') # default:facecolors=None
```
