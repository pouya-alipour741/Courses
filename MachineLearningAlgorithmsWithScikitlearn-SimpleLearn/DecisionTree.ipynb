```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
balance_data = pd.read_csv("data\Decision_Tree_ Dataset.csv", header=0, sep=",")
balance_data.head(10)
```

```python
balance_data[balance_data["Result"]=="No"].head(10)
```

```python
sorted_data = balance_data.sort_values(by=["Result"])
sorted_data
```

```python
sorted_data = sorted_data.drop(columns=["Sum"],axis=1)
sorted_data
```

```python
len(balance_data),balance_data.shape
```

```python
X = balance_data.values[:,:4]
X.shape
```

```python
y = balance_data.values[:,-1]
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=100)
X_train.shape
```

```python
clf_entropy = DecisionTreeClassifier(criterion="entropy",random_state=100,
                                    max_depth=3,min_samples_leaf=5)

clf_entropy.fit(X_train,y_train)
```

```python
y_pred = clf_entropy.predict(X_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```
