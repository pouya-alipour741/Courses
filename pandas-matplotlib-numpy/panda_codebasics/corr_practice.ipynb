```python
import numpy as np
def histogram_intersection(a, b):

    v = np.minimum(a, b).sum().round(decimals=1)

    return v
```

```python
f = pd.DataFrame([(.2, .3), (.0, .6), (.6, .0), (.2, .1)],

                  columns=['dogs', 'cats'])
f
```

```python
f.corr(method=histogram_intersection)
```
