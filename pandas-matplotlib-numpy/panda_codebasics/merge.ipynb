```python
import pandas as pd
import datetime as dt
```

```python
dateparser = lambda x:dt.datetime.strptime(x,"%Y-%m-%d %I-%p")
df = pd.read_csv("csv_data/ETH-1h.csv",parse_dates=["Date"],date_parser=dateparser,index_col="Date")
df
```

```python
test = df.loc[df.index >= "2020-1-3"]
test = test.copy()
test
```

```python
test["predictions"] = test["Close"]*1.1
test
```

```python
df
```

```python
pd.merge(df,test[["Volume","predictions"]],how="left")
```

```python
pd.merge(df,test[["Volume","predictions"]],how="left",left_index=True,right_index=True)
```

```python
df = pd.merge(df,test[["Volume","predictions"]],how="left",left_index=True,right_index=True).copy()
df
```
