```python
import pandas as pd
```

```python
df = pd.read_csv("Group_By/weather_by_cities.csv")
df
```

```python
###if all data in a row in duplicated in another row
df.duplicated()
```

```python
# if a value in a certain column is repeated
df.duplicated(subset=["city"])
```

```python
##locate every duplicate 
df.loc[df.duplicated(subset=["city"])]
```

```python
## ~df.duplicated means not the duplicates
df.loc[~df.duplicated(subset=["city"])]
```

```python
df.query('city == "new york"')
```

```python
df.query('city == "paris"')
```

```python
df.loc[df.duplicated(subset=["city","event"])]
```

```python
#dropped 2 rows 
df = df.loc[~df.duplicated(subset=["city","event"])]
df
```

```python
##let's fix the indexes 
df.reset_index(drop=True, inplace=True)
df
```

```python
ax = df["city"].value_counts()\
    .plot(kind="bar",title="title",figsize=(10,5))\
    
ax.set_xlabel("city")
ax.set_ylabel("count")

```
