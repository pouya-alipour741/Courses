```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x, y):
    return np.sin(x)**8 + np.cos(20+ y*x)*np.cos(y)
```

```python
x = np.linspace(0, 5, 5)
y = np.linspace(0, 5, 5)

xx, yy = np.meshgrid(x,y)
xx,yy
```

```python
zz = f(xx,yy)
zz
```

```python
plt.contour(xx,yy,zz)
plt.colorbar()
```

```python
plt.contour(xx,yy,zz, colors="black")
plt.colorbar()
```

```python
plt.contour(xx,yy,zz, cmap="BuPu", alpha=0.8, linestyles="dashed", linewidths=5)  #linestyles="dotted","dashdot"
plt.colorbar()
```

```python
plt.contourf(xx,yy,zz, cmap="BuPu", alpha=0.8)
plt.colorbar()
```
