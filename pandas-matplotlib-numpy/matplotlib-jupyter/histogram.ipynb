```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
b, w = 100,15

x = b + w*np.random.randn(10000)

fig = plt.figure(figsize=(15,10))
ax = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,3)
ax3 = fig.add_subplot(2,2,4)

ax.hist(x, bins=50, color="red", align="right");  ##orientation="left"
ax2.hist(x, bins=50, color="red", align="right", rwidth=0.6);
ax3.hist(x, bins=50, color="red", align="right", rwidth=0.6,histtype="step");
```

```python
fig0, ax = plt.subplots(figsize=(10,5))
fig0, ax2 = plt.subplots()
fig0, ax3 = plt.subplots()
fig0, ax4 = plt.subplots()
ax.hist(x, bins=50, color="red", align="right");  ##orientation="left"
ax2.hist(x, bins=50, color="red", align="right", rwidth=0.6);
ax3.hist(x, bins=50, color="red", align="right", rwidth=0.6,histtype="step");
ax4.hist(x, bins=50, color="red", align="right", rwidth=0.6,histtype="step", cumulative=True);
```

```python
fig, ax = plt.subplots(2,2,figsize=(15,10))
ax[0,0].hist(x, bins=50, color="red", align="right");  ##orientation="left"
ax[0,0].set_xlabel("X")
ax[0,1].hist(x, bins=50, color="red", align="right", rwidth=0.6);
ax[1,0].hist(x, bins=50, color="red", align="right", rwidth=0.6,histtype="step");
```
