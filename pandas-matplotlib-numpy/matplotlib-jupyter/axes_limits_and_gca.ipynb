```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(1,4,1000)
y = 2 + x**2

plt.plot(x,y)
```

```python
x[::100]
```

```python
plt.plot(x,y)
plt.xlim(0,7)
plt.ylim(0,22)
```

```python
plt.plot(x,y)
plt.ylim(0,None)
```

```python
plt.subplot(1,3,1)
plt.plot(x,y)


plt.subplot(1,3,2)
plt.scatter(x[::100],y[::100])
ax = plt.gca()

plt.subplot(1,3,3)
plt.plot(x,-y)

plt.tight_layout()
ax.set_xlim(0,10)
ax.set_ylim(0,None)
ax.set_title("scatter plot")
ax.set_xlabel("")
```

```python
plt.plot(x,y)
#reverse x label with static data
plt.xlim(4,1)
```

```python
#reverse x label for dynamic data
plt.plot(x,y)
left, right = plt.xlim()
print(left,right) #debug
plt.xlim(right,left)
```
