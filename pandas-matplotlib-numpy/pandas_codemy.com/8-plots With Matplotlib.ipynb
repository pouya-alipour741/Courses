```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline
```

```python
my_df = pd.DataFrame(randn(20,4),columns=['Monday','Tuesday',"Wendsday",'Thursday'])
my_df
```

```python
my_df["Wendsday"].hist(bins=10,grid=False,legend=True)
```

```python
my_df["Wendsday"].plot(kind='hist')
```

```python
##When stacked is True, each column must be either all positive or all negative. Column 'Monday' contains both positive and negative values
my_df.plot(kind='area',stacked=False)  
```

```python
my_df.abs().plot(kind='area',legend=False,alpha=0.5,title="My Awesome area plot",table=True)  ##absolute values
```

```python
my_df.abs().plot(kind='bar',legend=False,alpha=0.5,title='My Awesome area plot',stacked=True)
```

```python
my_df['Monday'].abs().plot(kind='bar',legend=False,alpha=0.5,title='My Awesome area plot')
```

```python
my_df.abs().plot.bar(grid=True,figsize=(10,5))
```

```python
my_df.plot(kind='line',figsize=(7,5),lw=2)  ##line width
```

```python
##https://matplotlib.org/stable/tutorials/colors/colormaps.html
my_df = pd.DataFrame(randn(500,4),columns=['Monday','Tuesday',"Wendsday",'Thursday'])
my_df.plot(kind='scatter',x='Monday',y="Tuesday",c='Wendsday',cmap="OrRd",s='Wendsday')
```

```python
my_df.plot(kind='scatter',x='Monday',y="Tuesday",c='Wendsday',cmap="OrRd",s=my_df['Wendsday']*200,alpha=0.5)
```

```python
my_df.plot(kind='box',legend=False,title='My Awesome area plot',grid=True)
```

```python
my_df.plot(kind='hexbin',x='Monday',y='Tuesday',gridsize=10,title='My awesome hexbin',grid=True)
```

```python
##second method
my_df.plot.hexbin(x='Monday',y='Tuesday',gridsize=10,title='My awesome hexbin',grid=True)
```
