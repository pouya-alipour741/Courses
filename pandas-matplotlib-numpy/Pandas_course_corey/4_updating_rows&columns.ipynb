```python
people = {
    "first": ["Corey", 'Jane', 'John'],
    "last": ["Schafer", 'Doe', 'Doe'],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com'],
}

import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
col_to_look_at = ["first","email"]
for col in df:
    if col in col_to_look_at:
        print(col)
```

```python
df.columns = ['first_name','last_name','email']  ##not so useful but you can change all column titles this way
df
```

```python
df.columns = [x.upper() for x in df.columns]  ##make column names all upper or lower
df
```

```python
df.columns = df.columns.str.replace('_',' ')
df
```

```python
df.rename(columns= {'FIRST NAME':'first','LAST NAME':'last'},inplace=True)
df
```

```python
df.loc[2] = ['pouya','alipour','a@b.com']
df
```

```python
df.loc[2,['first','last']] = ['loc method','testy']
df
```

```python
df.at[2,'first'] = 'at method'   ###you can also use .loc same as this
df
```

```python
df.loc[df['last'] == 'Schafer','first'] = 'hello'
df.loc[df['last'] == 'Schafer','last'] = 'hi'
df
```

```python
df['EMAIL'] = df['EMAIL'].str.lower()
df
```

```python
df.rename(columns={
    'EMAIL':'email',
}, inplace=True)
```
