```python
people = {
    "first": ["Corey", 'Jane', 'John','Adam'],
    "last": ["Schafer", 'Doe', 'Doe','Doe'],
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com','AdamDoe@email.com'],
}

import pandas as pd

df = pd.DataFrame(people)
df
```

```python
df.sort_values(by='last',ascending=False,inplace=True)
df
```

```python
df.sort_values(by=['last','first'],ascending=False,inplace=True)
df
```

```python
df.sort_values(by=['last','first'],ascending=[False,True],inplace=True)
df
```

```python
df.sort_index(inplace=True)
df
```

```python
df['last'].sort_values()         ##can also do it like this
df
```
