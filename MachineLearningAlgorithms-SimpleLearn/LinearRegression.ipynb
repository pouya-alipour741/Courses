```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
companies = pd.read_csv("data/1000_Companies.csv")
companies.head()
```

```python
companies
```

```python
X = companies.iloc[:,:-1]
X
```

```python
y = companies.iloc[:,4].values
y
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer([("State_transformed", ##just a name
                         OneHotEncoder(),
                         [3])], remainder = 'passthrough')
X = ct.fit_transform(X)                     
X
```

```python
##avoid dummy variable trap to remove the column names from data
X = X[:, 1:]
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
y_pred
```

```python
y_test
```

```python
# y = mx+c  now let's calculate c for each row
regressor.coef_
```

```python
regressor.intercept_
```

```python
##accuracy
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
